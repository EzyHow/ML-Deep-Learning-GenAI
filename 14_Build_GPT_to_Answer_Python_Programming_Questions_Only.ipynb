```python
import warnings
warnings.filterwarnings("ignore")

from tqdm.auto import tqdm
import matplotlib.pyplot as plt

# from transformers import AutoModelForCausalLM, AutoTokenizer
from transformers import pipeline, GPT2LMHeadModel, GPT2Tokenizer
```

# Load Model (GPT-2)

```python
# model_name = "openai-community/gpt2-medium"
# tokenizer = AutoTokenizer.from_pretrained(model_name)
# model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
# Load GPT-2
model_name = "openai-community/gpt2-medium"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)
```

```python
generator = pipeline("text-generation", model=model, tokenizer=tokenizer)
```

```python
def gpt2(prompt: str):
  output = generator(prompt, max_length=200, num_return_sequences=1, do_sample=True)
  return output[0]["generated_text"]
```

# Few Prompts Before Filtering

```python
prompts = [
    "GeeksForGeeks.org is a",
    "Once upon a time",
    "Python is used for"
]

for prompt in tqdm(prompts, desc="Generating text"):
    response = gpt2(prompt)

    print('\n\n=========================================\n\n')
    print('Prompt:\n', prompt)
    print('\n\nResponse:\n', response)
```

# Applying Filtering to Answer Only Python Programming Questions

```python
# A very simple filter function
def is_python_question(prompt: str) -> bool:
    keywords = ["python", "list", "dict", "tuple", "pandas", "numpy", "django",
                "flask", "py", "decorator", "class", "function", "def", "import"]
    return any(word.lower() in prompt.lower() for word in keywords)

def restricted_gpt2(prompt: str):
    if is_python_question(prompt):
        output = generator(prompt, max_length=200, num_return_sequences=1, do_sample=True)
        return output[0]["generated_text"]
    else:
        return "Sorry! I cannot answer this question."
```

# Testing Prompts After Applying Filtering

```python
prompts_for_testing = [
    "How do I create a list comprehension in Python?", # Python
    "What is the capital of France?", # Non-programming
    "Explain the concept of a decorator in Python.", # Python
    "Who wrote 'Romeo and Juliet'?", # Non-programming
    "How to define a class in Python?", # Python
    "What is the theory of relativity?", # Non-programming
    "How to use pandas to read a CSV file?", # Python
    "What are the steps to bake a cake?", # Non-programming
    "Explain different data types in Python.", # Python
    "What is the highest mountain in the world?" # Non-programming
]

for prompt in tqdm(prompts_for_testing, desc="Testing prompts"):
    response = restricted_gpt2(prompt)
    print('\n\n=========================================\n\n')
    print('Prompt:\n', prompt)
    print('\n\nResponse:\n', response)
```
